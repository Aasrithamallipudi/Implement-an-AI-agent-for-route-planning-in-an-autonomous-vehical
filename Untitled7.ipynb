```python
import numpy as np
import matplotlib.pyplot as plt
import heapq
import random
import time
from IPython.display import clear_output

GRID_SIZE = 20
STATIC_TRAFFIC_PROB = 0.1
ROADBLOCK_PROB = 0.05
NUM_MOVING_CARS = 10

NORMAL_COST = 1
STATIC_TRAFFIC_COST = 5
ROADBLOCK = -1

def create_grid():
    grid = np.zeros((GRID_SIZE, GRID_SIZE))
    for i in range(GRID_SIZE):
        for j in range(GRID_SIZE):
            rand = random.random()
            if rand < ROADBLOCK_PROB:
                grid[i, j] = ROADBLOCK
            elif rand < ROADBLOCK_PROB + STATIC_TRAFFIC_PROB:
                grid[i, j] = STATIC_TRAFFIC_COST
            else:
                grid[i, j] = NORMAL_COST
    return grid

class Node:
    def __init__(self, position, parent=None, g=0, h=0):
        self.position = position
        self.parent = parent
        self.g = g
        self.h = h

    def f(self):
        return self.g + self.h

    def __lt__(self, other):
        return self.f() < other.f()

def heuristic(a, b):
    return abs(a[0] - b[0]) + abs(a[1] - b[1])

def astar(grid, start, goal, moving_cars=None):
    open_set = []
    heapq.heappush(open_set, Node(start, None, 0, heuristic(start, goal)))
    closed_set = set()

    while open_set:
        current = heapq.heappop(open_set)
        if current.position == goal:
            path = []
            while current:
                path.append(current.position)
                current = current.parent
            return path[::-1]

        closed_set.add(current.position)

        neighbors = get_neighbors(current.position, grid)
        for neighbor_pos in neighbors:
            if neighbor_pos in closed_set:
                continue
            move_cost = grid[neighbor_pos]
            if move_cost == ROADBLOCK:
                continue
            if moving_cars and neighbor_pos in moving_cars:
                continue

            neighbor = Node(
                neighbor_pos,
                current,
                current.g + move_cost,
                heuristic(neighbor_pos, goal)
            )
            heapq.heappush(open_set, neighbor)
    return None

def get_neighbors(position, grid):
    neighbors = []
    dirs = [(-1, 0), (1, 0), (0, -1), (0, 1)]
    for d in dirs:
        nx, ny = position[0] + d[0], position[1] + d[1]
        if 0 <= nx < GRID_SIZE and 0 <= ny < GRID_SIZE:
            neighbors.append((nx, ny))
    return neighbors

def plot_grid(grid, path=None, start=None, goal=None, moving_cars=None):
    plt.figure(figsize=(10, 10))
    cmap = plt.cm.get_cmap('Set3', 4)

    display_grid = np.zeros_like(grid)
    for i in range(GRID_SIZE):
        for j in range(GRID_SIZE):
            if grid[i, j] == ROADBLOCK:
                display_grid[i, j] = 3
            elif grid[i, j] == STATIC_TRAFFIC_COST:
                display_grid[i, j] = 2
            else:
                display_grid[i, j] = 0

    plt.imshow(display_grid, cmap=cmap, origin="upper")

    if path:
        path_x, path_y = zip(*path)
        plt.plot(path_y, path_x, color="black", linewidth=2, label="Path")

    if moving_cars:
        cars_x, cars_y = zip(*moving_cars)
        plt.scatter(cars_y, cars_x, color="blue", s=100, marker="s", label="Moving Traffic")

    if start:
        plt.scatter(start[1], start[0], marker="o", color="green", s=200, label="Start")
    if goal:
        plt.scatter(goal[1], goal[0], marker="x", color="red", s=200, label="Goal")

    plt.legend()
    plt.title("Autonomous Vehicle Route Planning with Moving Traffic")
    plt.show()

def move_cars(moving_cars, grid):
    new_positions = set()
    for car in moving_cars:
        moves = get_neighbors(car, grid)
        if moves:
            move = random.choice(moves)
            if grid[move] != ROADBLOCK:
                new_positions.add(move)
            else:
                new_positions.add(car)
        else:
            new_positions.add(car)
    return list(new_positions)

def main():
    grid = create_grid()

    moving_cars = []
    while len(moving_cars) < NUM_MOVING_CARS:
        pos = (random.randint(0, GRID_SIZE - 1), random.randint(0, GRID_SIZE - 1))
        if grid[pos] != ROADBLOCK:
            moving_cars.append(pos)

    while True:
        start = (random.randint(0, GRID_SIZE - 1), random.randint(0, GRID_SIZE - 1))
        goal = (random.randint(0, GRID_SIZE - 1), random.randint(0, GRID_SIZE - 1))
        if start != goal and grid[start] != ROADBLOCK and grid[goal] != ROADBLOCK:
            break

    path = astar(grid, start, goal, moving_cars)

    steps = 0
    while path:
        clear_output(wait=True)

        if steps >= len(path):
            print("Goal Reached!")
            plot_grid(grid, path, path[-1], goal, moving_cars)
            break

        current_pos = path[steps]

        if current_pos in moving_cars:
            print("Blocked by moving car! Cannot proceed.")
            plot_grid(grid, path, current_pos, goal, moving_cars)
            break
        else:
            plot_grid(grid, path, current_pos, goal, moving_cars)
            steps += 1
            moving_cars = move_cars(moving_cars, grid)
            time.sleep(0.5)

if __name__ == "__main__":
    main()
```
